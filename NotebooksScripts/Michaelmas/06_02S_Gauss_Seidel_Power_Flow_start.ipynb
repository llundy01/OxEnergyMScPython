# 06_02 Solves a three bus system

```python
# Imports
import numpy as np
```

## Set up the constants of three bus system
Note, we only need to determine the voltages for Bus 2 and Bus 3


```python
SBase = 100  #to put things in per unit 
# Set up the S 

P2 = -256.6/SBase   #because its per 100MVA unit base
Q2 =-110.2/SBase
P3 = -138.6/SBase
Q3 = -45.2/SBase

# Set up the line admitances

y12 = 1/(0.02 +0.04j)
y13 = 1/(0.01 +0.03j)
y23 = 1/(0.0125 + 0.025j)

# Set up bus voltage of slack bus 1.
V1 = 1.05 +0j

# Initialise bus voltages.
V2_k = 1+0j
V3_k = 1+0j
```

```python
# Print out values
print("S")
print(f" S2 = {P2:.4f} + j {Q2:.4f} ")
print(f" S3 = {P3:.4f} + j {Q3:.4f} ")

print("\ny")
print(f" y12 = {y12:.4f} ")
print(f" y13 = {y13:.4f} ")
print(f" y23 = {y23:.4f} ")

print("\nV initial")
print(f" V1  = {V1:.4f} ")
print(f" V2_k  = {V2_k:.4f} ")
print(f" V3_k  = {V3_k:.4f} ")
```

## Equations needed to iterate load bus. 
See equation (5) in notes 06_02

```python
# Bus 2
V2_k1 = ( (P2-1j*Q2)/np.conj(V2_k) + y12*V1 + y23*V3_k )/(y12+y23)

# Bus 3
V3_k1 = ((P3-1j*Q3)/np.conj(V3_k) + y13*V1 + y23*V2_k)/(y13+y23)
```

```python
print(f"V2_k1 = {V2_k1:.4f}")
print(f"V3_k1 = {V3_k1:.4f}")
```

## Calculate the errors

```python
err2 = np.absolute(V2_k1-V2_k)
err3 = np.absolute(V3_k1-V3_k)
```

```python
print(f"err2 = {err2:.8f}")
print(f"err3 = {err3:.8f}")
```

## Make the new values equal to the old values for the next iteraion

```python
V2_k = V2_k1
V3_k = V3_k1
```

## This block puts it all together does the iteration 

```python
# Set up bus voltage of slack bus 1.
V1 = 1.05 +0j

# Initialise bus voltages.
V2_k = 1+0j
V3_k = 1+0j

# Set up the error criterion
eps = 1e-5

# Set the error so that the while loop will start
err2 = 1
err3 = 1

# Set up a counter to count the number of iterations
count = 0


while err2 > eps and err3 > eps:
    count += 1
# Bus 2
    V2_k1 = ( (P2-1j*Q2)/np.conj(V2_k) + y12*V1 + y23*V3_k )/(y12+y23)
# Bus 3
    V3_k1 = ((P3-1j*Q3)/np.conj(V3_k) + y13*V1 + y23*V2_k)/(y13+y23)
    
    err2 = np.absolute(V2_k1-V2_k)
    err3 = np.absolute(V3_k1-V3_k)
    
    print(f"{count:4} V2 = {V2_k1:.5f} V3 = {V3_k1:.5f}  err2 = {err2:9.4g} err3 = {err3:9.4g}")
  
    V2_k = V2_k1
    V3_k = V3_k1
```

## Determine all the currents and power flows

```python
# Slack bus P+jQ Use equation (4) 

S1 = np.conj(V1)*( V1*(y12+y13) - (y12*V2_k1 + y13*V3_k1) )*SBase
print(f"S1 = {S1:.4} ")
```

```python
# Determine the currents

I12 = y12*(V1 - V2_k1)
I21 = -I12

I13 = y13*(V1-V3_k1)
I31 = -I13

I23 = y23*(V2_k1 - V3_k1)
I32 = -I23

print(f"I12 = {I12:15.4} I21 = {I21:15.4}")
print(f"I13 = {I13:15.4} I31 = {I31:15.4}")
print(f"I23 = {I23:15.4} I21 = {I32:15.4}")
```

```python
# Power out of  bus i to bus j
S12 = V1*np.conj(I12)*SBase
S21 = V2_k1*np.conj(I21)*SBase

S13 = V1*np.conj(I13)*SBase
S31 = V3_k1*np.conj(I31)*SBase

S23 = V2_k1*np.conj(I23)*SBase
S32 = V3_k1*np.conj(I32)*SBase

print(f"S12 = {S12:15.4} S21 = {S21:15.4}")
print(f"S13 = {S13:15.4} S31 = {S31:15.4}")
print(f"S23 = {S23:15.4} S32 = {S32:15.4}")
```

```python
# Power absorbed in each line, see equation (10)
SL12 = S12 + S21
SL13 = S13 +S31
SL23 = S23+ S32

print(f"SL12 = {SL12:15.4} ")
print(f"SL13 = {SL13:15.4} ")
print(f"SL23 = {SL23:15.4} ")
```
