## Gauss Seidel Starter - I


```python
import numpy as np
import matplotlib.pyplot as plt
eps = 1e-6
maxIter = 100
```


Find x using GS given
$$x^2 + 3x + 1 = 0$$

```python
xold = 1
e = 1

count=0
while (e > eps)&(count<maxIter):
    count += 1
    # Insert equations here:
    xnew = -(xold**2 +1)/3
    #
    e = np.absolute(xnew - xold)
    xold = xnew
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"Answer = {xnew}. Interations = {count}")
```

## Check: Plot function

```python
x = np.linspace(start=-2, stop=2, num=50)
plt.plot(x,x**2+3*x+1)
plt.grid()
```

## Gauss Seidel Starter - II

Find x using GS given
$$x^3-6x^2+9x-3= 0$$

```python
xold = 1
e = 1
count=0
while (e > eps)&(count<maxIter):
    count += 1
    # Insert equations here:
    xnew = (3-xold**3+6*xold**2)/9
    #
    e = np.absolute(xnew - xold)
    xold = xnew
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"Answer = {xnew}. Interations = {count}")
```

## Check: Plot function

```python
x = np.linspace(start=-0, stop=5, num=50)
plt.plot(x,x**3-6*x**2+9*x-3)
plt.grid()
```

## Gauss Seidel Starter - III

Find x using GS given
$$x^3-6x^2+9x-3= 0$$
$$y^2 -x = 0$$

```python
xold = 2
yold = 2
e=1
count=0
while (e > eps)&(count<maxIter):
    # Insert equations here:
    xnew = (3-xold**3+6*xold**2)/9
    ynew = yold**2/xnew
    #
    e = max( np.absolute(xnew - xold),  np.absolute(ynew - yold))
    xold = xnew
    yold = ynew
    count =count +1
    
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"xnew = {xnew} ynew = {ynew} Error = {e:.4e} Count = {count}")
```
