```python
import math 
import pandas as pd



def PV_prod(roof_length, roof_width, irradiance_kwh_per_m2, gap_mm):
    """
    Retourne pour chaque modèle de 'panels' :
      - nombre de panneaux pouvant être installés (grid packing, meilleure orientation),
      - surface totale occupée,
      - production mensuelle (kWh) pour la totalité de l'installation,
      - production mensuelle par panneau,
      - production annuelle totale et par panneau,
      - coût estimé total (si cost présent).
    Arguments :
      roof_length, roof_width : dimensions du toit en mètres
      irradiance_kwh_per_m2 : liste de 12 valeurs (kWh/m2 par mois, Jan..Dec)
      gap_mm : espacement entre panneaux en mm (appliqué en L et W)
    Retour :
      (results_list, df_sorted)  -- results_list = liste de dicts détaillés, df_sorted = pandas.DataFrame trié par production annuelle (desc)
    """
    # paramètres par défaut utilisables (si tu veux les changer plus tard)
    performance_ratio = 0.80

    # validation entrée irradiance
    if not (isinstance(irradiance_kwh_per_m2, (list, tuple)) and len(irradiance_kwh_per_m2) == 12):
        raise ValueError("irradiance_kwh_per_m2 must be a list/tuple of 12 monthly values (Jan..Dec)")

    # fonction utilitaire : calcule combien de panneaux tiennent en grille, en testant les 2 orientations
    def nb_panels(roof_length, roof_width, panel_length, panel_width, gap_mm):
        gap_m = gap_mm / 1000.0
        def best_orientation(panel_length_local, panel_width_local):
            fp_length = panel_length_local + gap_m
            fp_width = panel_width_local + gap_m
            length_axis = math.floor((roof_length - 2*gap_m)/ fp_length)
            width_axis  = math.floor((roof_width- 2*gap_m)/ fp_width)
            return max(0, int(length_axis) * int(width_axis)) #taking into account the two sides of the roof
        return max(best_orientation(panel_length, panel_width),
                   best_orientation(panel_width, panel_length))
        ## we calculate for one side of the roof but there will be two sides 

    months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

    results = []

    for p in panels:
        # unpack minimalement (on garde ta logique existante mais safe)
        model = p[0]
        manu  = p[1] if len(p) > 1 else ""
        Lmm   = p[2] if len(p) > 2 else 0
        Wmm   = p[3] if len(p) > 3 else 0
        nominal_Wp = p[4] if len(p) > 4 else None
        eff_pct    = p[5] if len(p) > 5 else None
        cost_per_panel = p[6] if len(p) > 6 else 0.0

        # dimensions en m
        L = Lmm / 1000.0
        W = Wmm / 1000.0

        # combien de panneaux tiennent sur le toit (meilleure orientation)
        n_panels = nb_panels(roof_length, roof_width, L, W, gap_mm)

        panel_area = round(L * W, 6)    # m2
        total_area = round(panel_area * n_panels, 6)

        # fallback efficiency si absent
        eff = eff_pct if (eff_pct is not None and eff_pct > 0) else 20.0

        # calcul mensuel (par panneau et par array)
        monthly_kwh_per_panel = []
        monthly_kwh_array = []
        total_capacity = n_panels*nominal_Wp
        for irr in irradiance_kwh_per_m2:
            # production = irr (kWh/m2/mo) * area (m2) * efficiency * performance_ratio
            kwh_panel = irr * panel_area * (eff / 100.0) * performance_ratio
            kwh_array = kwh_panel * n_panels
            monthly_kwh_per_panel.append(round(kwh_panel, 4))
            monthly_kwh_array.append(round(kwh_array, 3))

        annual_kwh_array = round(sum(monthly_kwh_array), 3)
        annual_kwh_per_panel = round(sum(monthly_kwh_per_panel), 6)
        est_cost_total = round(cost_per_panel * n_panels, 2)

        results.append({
            "model": model,
            "manufacturer": manu,
            "panel_length_m": L,
            "panel_width_m": W,
            "panel_area_m2": panel_area,
            "panels_fit": n_panels,
            "total_area_m2": total_area,
            "total_capacity_W" : total_capacity,
            "nominal_Wp": nominal_Wp,
            "eff_pct": eff,
            "monthly_kwh_per_panel": monthly_kwh_per_panel,
            "monthly_kwh_array": monthly_kwh_array,
            "annual_kwh_array": annual_kwh_array,
            "annual_kwh_per_panel": annual_kwh_per_panel,
            "est_cost_total_gbp": est_cost_total
        })

    # construire un DataFrame lisible et trier par production annuelle décroissante
    rows = []
    for r in results:
        row = {
            "model": r["model"],
            "manufacturer": r["manufacturer"],
            "panels_fit": r["panels_fit"],
            "total_capacity_W" : r["total_capacity_W"],
            "total_area_m2": r["total_area_m2"],
            " est_cost_total_gbp": r["est_cost_total_gbp"],
            "annual_kwh_array": r["annual_kwh_array"]
        }
        for mname, val in zip(months, r["monthly_kwh_array"]):
            row[mname] = val
        rows.append(row)

    df = pd.DataFrame(rows)
    df_sorted = df.sort_values("annual_kwh_array", ascending=False).reset_index(drop=True)

    return results, df_sorted
```

```python
#for Jinko Solar


import pandas as pd

df = pd.read_csv(
    "ninja_pv_51.7501_-1.2738_corrected.csv",
    comment='#',
    parse_dates=['local_time']
)

# Calcul de la production mensuelle (kWh)
monthly_kwh = df.resample('M', on='local_time')['electricity'].sum().round(2)

# Conversion en DataFrame lisible
monthly_kwh_df = monthly_kwh.reset_index()
monthly_kwh_df['Month'] = monthly_kwh_df['local_time'].dt.strftime('%b')
monthly_kwh_df = monthly_kwh_df[['Month', 'electricity']]
monthly_kwh_df.columns = ['Month', 'Production_kWh']

# Export en CSV
monthly_kwh_df.to_csv("production_mensuelle_pv.csv", index=False)

print("✅ Fichier 'production_mensuelle_pv.csv' créé avec succès !")
```
